```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
import numpy as np
from numpy import asarray, save
from os import listdir
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
```

```python
from matplotlib.image import imread
# define location of dataset
folder = 'Dataset/train/'
# plot first few images
for i in range(9):
	# define subplot
	plt.subplot(330 + 1 + i)
	# define filename
	filename = folder + 'Smile-' + str(i) + '.png'
	# load image pixels
	image = imread(filename)
	# plot raw pixel data
	plt.imshow(image)
# show the figure
plt.show()
```

```python
folder = 'Dataset/train/'
photos, labels = list(),list()

for file in listdir(folder):
    output=0.0
    if file.startswith('Smile'):
        output =1.0
    photo = load_img(folder + file, target_size=(200,200))
    photo=img_to_array(photo)
    
    photos.append(photo)
    labels.append(output)

photos = asarray(photos)
labels=asarray(labels)
print(photos.shape, labels.shape)

save('train_smiley_frown_photos.npy', photos)
save('train_smiley_frown_labels.npy', labels)
```

```python
folder = 'Dataset/test/'
photos, labels = list(),list()

for file in listdir(folder):
    output=0.0
    if file.startswith('Smile'):
        output =1.0
    photo = load_img(folder + file, target_size=(200,200))
    photo=img_to_array(photo)
    
    photos.append(photo)
    labels.append(output)

photos = asarray(photos)
labels=asarray(labels)
print(photos.shape, labels.shape)

save('test_smiley_frown_photos.npy', photos)
save('test_smiley_frown_labels.npy', labels)
```
